```python
import numpy as np
import pandas as pd
import os
```

```python
df_list_scenario = []


for file_name in os.listdir("./result_scenario"):
    file_to_open = "./result_scenario/%s" % file_name
    df_list_scenario.append(pd.read_csv(file_to_open))
```

```python
df_summary = pd.read_csv("./data/summary.csv")
```

```python
df_summary
```

```python
df_summary.columns = range(df_summary.shape[1])
```

```python
list_df_first_process = []
list_df_second_process = []
list_df_third_process = []
list_df_fourth_process = []
list_df_fifth_process = []

for df in df_list_scenario:
    
    df.columns = range(df.shape[1])
    
    df_first_process = df[range(0,3)]
    df_second_process = df[range(4,7)]
    df_third_process = df[range(8,11)]
    df_fourth_process = df[range(12,15)]
    df_fifth_process = df[range(16,19)]
    
    # reindexing
    df_first_process.columns = range(df_first_process.shape[1])
    df_second_process.columns = range(df_second_process.shape[1])
    df_third_process.columns = range(df_third_process.shape[1])
    df_fourth_process.columns = range(df_fourth_process.shape[1])
    df_fifth_process.columns = range(df_fifth_process.shape[1])
    
    list_df_first_process.append(df_first_process)
    list_df_second_process.append(df_second_process)
    list_df_third_process.append(df_third_process)
    list_df_fourth_process.append(df_fourth_process)
    list_df_fifth_process.append(df_fifth_process)
```

```python
list_scenario = []

for cnt in range(len(df_list_scenario)):
    
    list_answer = []
    
    for index, row in df_summary.iterrows():
        # First column as the process
        process_id = row[1]

        df_choose = None

        if process_id == 1:
            df_choose = list_df_first_process[cnt]
        elif process_id == 2:
            df_choose = list_df_second_process[cnt]
        elif process_id == 3:
            df_choose = list_df_third_process[cnt]
        elif process_id == 4:
            df_choose = list_df_fourth_process[cnt]
        else:
            df_choose = list_df_fifth_process[cnt]

        filter_time_diff = row[2]

        answer = np.max(df_choose[df_choose[2] == filter_time_diff][0])
        list_answer.append(answer)
    
    list_scenario.append(list_answer)
```

```python
list_scenario
```

```python
final_result = pd.DataFrame()


for counter in range(len(list_scenario)-1, -1, -1):
    
    number_row = df_summary[0].shape[0]
    
    result = pd.concat([pd.DataFrame(np.repeat(counter+1, number_row), columns=['Scenario']),
                        pd.DataFrame(df_summary[0].as_matrix(), columns=['Parts']),
                        pd.DataFrame(df_summary[1].as_matrix(), columns=['Process']),
                        pd.DataFrame(df_summary[2].as_matrix(), columns=['Process Time']),
                        pd.DataFrame(list_scenario[counter], columns=['Time'])], axis=1)
    
    final_result = pd.concat([result, final_result])
```

```python
final_result.to_csv("all_result.csv", index=False)
```
